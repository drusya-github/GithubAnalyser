```python
#saving in excel sheet

xl = openpyxl.Workbook()
sheet = xl.active
sheet.title = 'Top rated tvshows'
sheet.append(['Project Name','user_name','description','Programming Language','No. Of Stasrs overall','No.of stars Today'])
```

```python
#maincode

try:
    source = requests.get("https://github.com/trending")
    source.raise_for_status()
    soup = BeautifulSoup(source.text, 'html.parser')
    print(soup)
    
    for projects in trending:
        project_name = projects.find('h2', class_="h3 lh-condensed").a.get_text(strip=True).split("/")[-1]
        user_name = projects.find('h2', class_="h3 lh-condensed").a.get_text(strip=True).split("/")[0]
    
        p_tag = projects.find('p', class_="col-9 color-fg-muted my-1 pr-4")
        if p_tag is not None:
            # Use a list comprehension to extract only the text from the p tag's direct descendants
            description = ''.join([str(child) for child in p_tag.contents if isinstance(child, str)]).strip()
        else:
            print("Description not found for project")
    
        prog_lang = projects.find("div",class_= "f6 color-fg-muted mt-2").span.get_text(strip=True)
    
        stars = projects.find("a", class_="Link--muted d-inline-block mr-3").get_text(strip=True)
    
        stars_today = projects.find("div", class_="f6 color-fg-muted mt-2").get_text(strip=True).split("by")[1]
        stars_todayy = stars_today.split("stars today")[0]
    
        print(project_name, user_name, description, prog_lang, stars, stars_todayy)
        sheet.append([project_name, user_name, description, prog_lang, stars, stars_todayy])
except Exception as e:
    print(e)

xl.save('Github Stats.xlsx')
```
